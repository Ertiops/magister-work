```python
import numpy as np
import pandas as pd
import nltk
import re
import os
import codecs
from sklearn import feature_extraction
import mpld3
import pickle
```

```python
with open('phys_corpus.pickle', 'rb') as handle:
    corpus = pickle.load(handle)
```

```python
def text_normalizator(corpus):
    for key in corpus.keys(): 
        corpus[key] = re.sub('\d+', '', corpus[key])
        corpus[key] =  re.sub('\\t|\\n|\+', '', corpus[key])
        corpus[key] =  re.sub('[^ЁёА-я\s]', '', corpus[key])
        corpus[key] =  re.sub('\s+', ' ', corpus[key])
    return corpus
```

```python
cleaned_corpus = corpus
```

```python
text_normalizator(cleaned_corpus)
```
