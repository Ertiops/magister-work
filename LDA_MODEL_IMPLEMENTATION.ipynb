```python
def get_list_of_lists(list_of_tuples):
    list_of_lists = []                                                          
    for tuple in list_of_tuples:
        list_of_lists.append(list(tuple))

    return list_of_lists
```

```python
titles_probabilities_to_dict = {}
for key in titles_probabilities:
    if key[0] not in titles_probabilities_to_dict:
        titles_probabilities_to_dict[key[0]] = get_list_of_lists(key[1])      
```

```python
titles_probabilities_to_dict
```

```python
def max_probable_topic(probs):
    titles_in_topics = {}
    for i in probs:
        titles_in_topics[i[0]] = i[1]
    if len(probs) != 1:
        max_value = max(titles_in_topics.values())
        for i in titles_in_topics:
            if titles_in_topics[i] == max_value:
                return i
    elif len(probs) == 1:
        return probs[0][0]
```

```python
topics_and_titles = {}

for key in titles_probabilities_to_dict:
    probs = titles_probabilities_to_dict[key]
    topic = max_probable_topic(probs)
    if topic not in topics_and_titles:
        topics_and_titles[topic] = [key]
    elif topic in topics_and_titles:
        topics_and_titles[topic].append(key)
```

```python
topics_and_titles
```

```python
for i in topics_and_titles:
    title_txt = f'themes_of_topic_{i}.txt'
    file = open(title_txt, 'w', encoding = 'utf-8')
    for j in topics_and_titles[i]:
        file.write(j + '\n')
    file.close
```
